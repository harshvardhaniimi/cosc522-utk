```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import math as math
import scipy.stats as sci
```

## For Class 1 and 2

```python
# initialising mu for class 1
mu_1 = [1, 1.375]

# initialising sigma^2 for class 1
sigma_1 = [ [ 0 for i in range(2) ] for j in range(2)]
sigma_1[0][0] = 0.33
sigma_1[0][1] = 0.0183
sigma_1[1][0] = 0.0183
sigma_1[1][1] = 0.0158

# initialising mu for class 2
mu_2 = [0.7, 1.05]

# initialising sigma^2 for class 2
sigma_2 = [ [ 0 for i in range(2) ] for j in range(2)]
sigma_2[0][0] = 0.0083
sigma_2[0][1] = 0.0075
sigma_2[1][0] = 0.0075
sigma_2[1][1] = 0.009
```

```python
# The 2-dimensional distribution is over variables X and Y
N = 100
X = np.linspace(-0.5, 2.5, N)
Y = np.linspace(0.8, 2, N)
X, Y = np.meshgrid(X, Y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
```

```python
# Defining the function to calculate PDF
def gaussian_pdf(pos, mu, sigma):
    sigma_inv = np.linalg.inv(sigma)
    nf = len(mu)
    dr = np.sqrt((2*np.pi)** nf * np.linalg.det(sigma))
    l = np.exp(-np.einsum('...k,kl,...l->...', pos-mu, sigma_inv, pos-mu)/2)/dr
    return l
```

```python
Z1 = gaussian_pdf(pos, mu_1, sigma_1)
Z2 = gaussian_pdf(pos, mu_2, sigma_2)
```

```python
fig = plt.figure()
ax = fig.gca(projection = "3d")
ax.plot_surface(X, Y, Z1, rstride=3, cstride=3, linewidth=1, antialiased=True)
plt.show()
```

```python
plt.gca(projection = "3d").plot_surface(X, Y, Z1, rstride=3, cstride=3, linewidth=1, antialiased=True)
```

```python
fig = plt.figure()
plt.contour(X, Y, Z1)
plt.contour(X, Y, Z2)
plt.plot(0.85, 1.15, "ro")
plt.show()
plt.savefig('plot1b.png')
```
