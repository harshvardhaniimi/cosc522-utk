```python
# differential function
def dft(x):
    return(50*np.cos(x)+2*x)
```

```python
# gradient descent algorithm
def minima(start_x, dft, rate = 0.01, precision = 0.001, max_i = 1000):
    i = 0 #iterations counting
    previous_step_size = 1 # initial step size
    
    opt = []
    
    while (previous_step_size > precision) and (i < max_i):
        opt.append(start_x)
        last_x = start_x #Store current x value in prev_x
        start_x = start_x - rate * dft(last_x) #Grad descent formula
        previous_step_size = abs(start_x - last_x) #Updating step size
        i = i + 1 #iteration count
        print("Iteration",i,"gives optimal minima as: ",start_x) #Print iterations
    print("Minima occurs at: ", start_x)
    return start_x, opt
```

```python
m_7 = minima(7, dft, rate = 0.05)
```

```python
def perceptron(X, Y, max_i = 10):
    w = np.random.uniform(0,1,2)
    b = np.random.uniform(0,1,1)
    for i in range(max_i):
        mismatch = 0
        for j in range(X.shape[0]):
            yhat = predict(X[j,:], w, b)
            if (yhat != Y[j]):
                mismatch = mismatch + 1
                w = w + X[j,:]*(Y[j]-yhat)
                b = b - (Y[j]- yhat)
                print(w, b)
            print(f"Epoch: {i} Iter: {j} Input: {X[j,:]} Y: {Y[j]} Y_hat: {yhat}")
        if(mismatch == 0):
            break
    return w, b
```

```python
# step function
def step(x):
    if x>= 0:
        return 1
    else:
        return 0
# predict
def predict(x, w, b):
    u = np.dot(w, x) - b
    y = step(u)
    
    return y
```

```python
# OR Gate
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([0,1,1,1])
```
