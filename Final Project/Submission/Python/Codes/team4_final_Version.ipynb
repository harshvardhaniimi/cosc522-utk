## For standardized dataset (nX)

```python
# standardized data 'nx'
feat_nX_tr, labels_tr, feat_nX_te, labels_te = load_heart(test_pct=0)
P0 = 1 - mean(labels_tr)
```

```python
# Case 1 - Euclidean distance classifier
mindist = MinDistClassifier(P0=P0)

# Case 2 - Mahalanobis distance classifier
mah = MahalanobisClassifier(P0=P0)

# Case 3 - Quadratic classifier 
quad = QuadraticClassifier(P0=P0)

#kNN is below

# BPNN
bpnn = NeuralNetwork()

# SVM using sklearn
svm = SVC(kernel='poly')

# Random forest using sklearn
rf = RandomForestClassifier(max_depth=2, random_state=0)

# Passive aggressive classifier using sklearn
passive_aggro = sklearn.linear_model.PassiveAggressiveClassifier()
```

```python
# BKS fusion
bks = BKS((mah, quad))

# Naive Bayes fusion
nb = NaiveBayes((mah, quad))
```

### Testing for nX

```python
# Case 1 - Euclidean distance classifier
print("Case 1 - Euclidean distance classifier")
mindist.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(mindist, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')

# Case 2 - Mahalanobis distance classifier
print("Case 2 - Mahalanobis distance classifier")
mah.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(mah, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')

# Case 3 - Quadratic classifier 
print("Case 3 - Quadratic classifier ")
quad.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(quad, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')


# kNN classifier 
print('kNN classifier')
ov_acc = []
k_values = []
for i in range(1, int(floor(sqrt(len(labels_tr)))) + 1):
    knn = KnnClassifier(k=i)
    knn.fit(feat_nX_tr, labels_tr)
    print(cross_val_acc(knn, feat_nX_tr, labels_tr, k=5))
    k_values.append(i)
    ov_acc.append(cross_val_acc(knn, feat_nX_tr, labels_tr, k=5)[0])    

plt.plot(k_values, ov_acc, c='g')
plt.xlabel('k Values')
plt.ylabel('Overall accuracy')
plt.title('Comparison of k values and accuracies for nX')
plt.legend()
plt.show()
print('----------------------------------------------------------------')



# BPNN
print("BPNN")
bpnn.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(bpnn, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')

# SVM using sklearn
print("SVM using sklearn")
svm.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(svm, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')

# Random forest using sklearn
print("Random forest using sklearn")
rf.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(rf, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')

# Passive aggressive classifier using sklearn
print("Passive aggressive classifier using sklearn")
passive_aggro.fit(feat_nX_tr, labels_tr)
print(cross_val_acc(passive_aggro, feat_nX_tr, labels_tr, k=5))
print('----------------------------------------------------------------')
```

```python
class NeuralNetwork1():

    def __init__(self, layers=[12, 256, 2], mini_batch_size=10, eta=0.08, 
                 max_epochs=1001, early_stop_window=50):
        self.layers = layers

        self._net = Network(self.layers)
        self._mini_batch_size = mini_batch_size
        self._eta = eta
        self._max_epochs = max_epochs
        self._early_stop_window = early_stop_window

    def fit(self, X, y):

        training_data, validation_data, _, _ = load_data_wrapper(test_pct=50, 
                                                                 supplied_data=(X, y,
                                                                                [], [],
                                                                                ),
                                                                 )

        self._net.SGD(training_data, epochs=self._max_epochs, 
                      mini_batch_size=self._mini_batch_size, 
                      eta=self._eta, test_data=validation_data, 
                      memory=self._early_stop_window)

    def predict(self, X):
        return array([np.argmax(self._net.feedforward(x[:,None])) for x in X])
```

```python
# BPNN
print("BPNN")
bpnn1 = NeuralNetwork()

bpnn1.layers[0] = fX_data.shape[1]
bpnn1.fit(fX_data, labels_te)
print(cross_val_acc(bpnn1, fX_data, labels_te, k=5))
'''
bpnn.fit(fX_data, labels_te)
print(cross_val_acc(bpnn, fX_data, labels_te, k=5))
print('----------------------------------------------------------------')
'''
```
